```python
#These are the packages for this project
#This project is done in Python 2
import pandas_datareader
import pandas as pd
import datetime
import pandas_datareader.data as web
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Timeframe I want to analyze
start = datetime.datetime(2016,1,1)
end = datetime.datetime(2017,1,1)
```

```python
#Starting of with Netflix
Netflix = web.DataReader('NFLX', 'google',start,end)
```

```python
Netflix.head()
```

```python
# Repeat the same process for Ebay, Amazon, and Alibaba
Ebay = web.DataReader('EBAY','google',start,end)
Amazon = web.DataReader('AMZN','google',start,end)
Alibaba = web.DataReader('BABA','google',start,end)
```

```python
Ebay.head()
```

```python
Amazon.head()
```

```python
Alibaba.head()
```

```python
#Creating visual of the opening values of the 4 stocks
Netflix['Open'].plot(label='NLFX',figsize=(16,8),title='Open Prices')
Ebay['Open'].plot(label='EBAY')
Amazon['Open'].plot(label='AMZN')
Alibaba['Open'].plot(label='BABA')
plt.legend();
```

```python
#Look and compare at the volume traded daily within the timeframe of each company
Netflix['Volume'].plot(label='NFLX',figsize=(16,8),title='Volume Traded')
Ebay['Volume'].plot(label='EBAY')
Amazon['Volume'].plot(label='AMZN')
Alibaba['Volume'].plot(label='BABA')
plt.legend();
```

```python
#Creating a volume to calculate the total amount of shares traded per day
#Total Traded = volume * open
Netflix['Total Traded'] = Netflix['Open']*Netflix['Volume']
Ebay['Total Traded'] = Ebay['Open']*Ebay['Volume']
Amazon['Total Traded'] = Amazon['Open']*Amazon['Volume']
Alibaba['Total Traded'] = Alibaba['Open']*Alibaba['Volume']
```

```python
#This graph will show the total amount of shares being traded daily
Netflix['Total Traded'].plot(label='NFLX',figsize=(16,8))
Ebay['Total Traded'].plot(label='EBAY')
Amazon['Total Traded'].plot(label='AMZN')
Alibaba['Total Traded'].plot(label='BABA')
plt.legend()
plt.ylabel('Total Traded')
```

```python
#The smooth moving averages for Netflix
Netflix['MA50'] = Netflix['Open'].rolling(50).mean()
Netflix['MA200'] = Netflix['Open'].rolling(200).mean()
Netflix[['Open','MA50','MA200']].plot(label='NFLIX',figsize=(14,8))
```

```python
#The smooth moving averages for Ebay
Ebay['MA50'] = Netflix['Open'].rolling(50).mean()
Ebay['MA200'] = Netflix['Open'].rolling(200).mean()
Ebay[['Open','MA50','MA200']].plot(label='EBAY',figsize=(14,8))
```

```python
#The smooth moving averages for Amazon
Amazon['MA50'] = Netflix['Open'].rolling(50).mean()
Amazon['MA200'] = Netflix['Open'].rolling(200).mean()
Amazon[['Open','MA50','MA200']].plot(label='AMZN',figsize=(14,8))
```

```python
#The smooth moving averages for Alibaba
Alibaba['MA50'] = Alibaba['Open'].rolling(50).mean()
Alibaba['MA200'] = Alibaba['Open'].rolling(200).mean()
Alibaba[['Open','MA50','MA200']].plot(label='BABA',figsize=(14,8))
```

```python
#Calculating the Percentage of Daily Returns in each stock
Netflix['returns'] = (Netflix['Close'] / Netflix['Close'].shift(1) ) - 1
Ebay['returns'] = (Ebay['Close'] / Ebay['Close'].shift(1) ) - 1
Amazon['returns'] = (Amazon['Close'] / Amazon['Close'].shift(1) ) - 1
Alibaba['returns'] = (Alibaba['Close'] / Alibaba['Close'].shift(1) ) - 1
```

```python
#Comparing the distribution of each stock
Netflix['returns'].plot(kind='kde',label='Netflix',figsize=(12,6))
Ebay['returns'].plot(kind='kde',label='EBAY')
Amazon['returns'].plot(kind='kde',label='AMZN')
Alibaba['returns'].plot(kind='kde',label='BABA')
plt.legend()
```

```python
box_df = pd.concat([Netflix['returns'],Ebay['returns'],Amazon['returns'],Alibaba['returns']],axis=1)
box_df.columns = ['Netflix Returns',' Ebay Returns','Amazon Returns', 'Alibaba Returns']
box_df.plot(kind='box',figsize=(8,11),colormap='jet')
```

```python
#Culumative Returns on each stock
Netflix['Cumulative Return'] = (1 + Netflix['returns']).cumprod()
Ebay['Cumulative Return'] = (1 + Ebay['returns']).cumprod()
Amazon['Cumulative Return'] = (1 + Amazon['returns']).cumprod()
Alibaba['Cumulative Return'] = (1 + Alibaba['returns']).cumprod()
```
